## 2. MAT File

.mat extension files contain MATLAB formatted data. scipy.io.loadmat() function reads .mat files and returns data in form of dict. 

```python
import scipy.io
import numpy as np

data = scipy.io.loadmat("heightWeight.mat")
```

```python
data.values()
```

The below cell converts .mat file to .csv file

```python
for i in data:
    if '__' not in i and 'readme' not in i:
        np.savetxt(("new"+i+".csv"),data[i],delimiter=',')
```

```python
df = pd.read_csv("newheightWeightData.csv",header=None)
```

```python
df.head()
```

## 3.DAT File 

DAT file is a generic data file which consists of data in form of text, binary, video etc. As the data seperator for this file is white space, regular expression '\s+' is specified as the seperator argument. 

```python
dat_df= pd.read_csv('data.dat', sep='\s+', header=None)
```
