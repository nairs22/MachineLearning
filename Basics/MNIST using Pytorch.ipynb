```
import torch
from matplotlib import pyplot as plt
import numpy as np
from torchvision import datasets, transforms
```

```
transform = transforms.Compose([transforms.Resize((28,28)),      
                               transforms.ToTensor(),
                               transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
                               ])
training_data=datasets.MNIST(root='/data',train=True,download=True,transform=transform)   # train parameter will set this dataset as training dataset. thransform parameter dictates any transformation you wich to apply on the images
```

transforms.compose will transform the image to a tensor. Here the transforms.ToTensor will transform from numpy 0,255 to float tensor values. transform.Normalise normalises the height and width of the image. the first bracket describes the mean and the second is the standard deviation.
